3. Explain the following terms
.

The difference in np.array, np.asarray and np.asanyarray

```python
#np.array always creates a copy of the input, even if it's already an array.
#np.asarray converts the input to an array only if it's not already an array. It doesn't create a copy if the input is already an array.
#np.asanyarray is similar to np.asarray, but it passes ndarray subclasses through. This means that if the input is a subclass of ndarray (like a masked array), it won't be converted to a regular ndarray.
```

The difference between Deep copy and shallow copyy

```python
#shallow coppy: shallow copy will change in one array and it leads to change in other.


arr=([1,2,3,4,5])
a=arr
a[0]=10
print( f"after changin in one it will reflect in both {arr}")
print(f"after changin in one it will reflect in both {a}")
```

```python
#deep copy: deep copy will change inn one not in others
arr=([1,2,3,4,5])
a=arr.copy()
a[0]=10
print( f"it wont reflect in both {arr}")
print(f"it will reflect only in one {a}")
```


15) Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and Perform the following tasks usingMatplotlib

a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and right y-axis for 'Humidity').

b) Label the x-axis as 'Date'.

c) Set the title of the plot as 'Temperature and Humidity Over Time'.

```python

import numpy as np
import matplotlib.pyplot as plt

# Create a sample time-series dataset
dates = pd.date_range(start='2023-01-01', periods=100, freq='D')
temperature = np.random.randint(20, 35, size=100)
humidity = np.random.randint(40, 80, size=100)

# Create a DataFrame
df = pd.DataFrame({'Date': dates, 'Temperature': temperature, 'Humidity': humidity})

# Plot Temperature and Humidity
fig, ax1 = plt.subplots(figsize=(10, 6))

# Temperature (left y-axis)
ax1.plot(df['Date'], df['Temperature'], color='tab:red', label='Temperature')
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (°C)', color='tab:red')
ax1.tick_params(axis='y', labelcolor='tab:red')

# Humidity (right y-axis)
ax2 = ax1.twinx()
ax2.plot(df['Date'], df['Humidity'], color='tab:blue', label='Humidity')
ax2.set_ylabel('Humidity (%)', color='tab:blue')
ax2.tick_params(axis='y', labelcolor='tab:blue')

# Title
plt.title('Temperature and Humidity Over Time')

# Show the plot
plt.show()
```

17 Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot


```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate random data
x = np.random.randn(100)
y = np.random.randn(100)

# Create a DataFrame for Seaborn
df = pd.DataFrame({'x': x, 'y': y})

# Assign quadrants
df['quadrant'] = 0  # Initialize
df.loc[(df['x'] >= 0) & (df['y'] >= 0), 'quadrant'] = 1
df.loc[(df['x'] < 0) & (df['y'] >= 0), 'quadrant'] = 2
df.loc[(df['x'] < 0) & (df['y'] < 0), 'quadrant'] = 3
df.loc[(df['x'] >= 0) & (df['y'] < 0), 'quadrant'] = 4

# Create scatter plot
sns.scatterplot(x='x', y='y', hue='quadrant', data=df, palette='viridis')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Quadrant-wise Scatter Plot")
plt.legend()
plt.show()
```

18 With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine Wave Function'

```python
import numpy as np
from bokeh.plotting import figure, show

# Generate x and y values for the sine wave
x = np.linspace(0, 2*np.pi, 1000)
y = np.sin(x)

# Create the Bokeh figure
p = figure(title="Sine Wave Function", x_axis_label='x', y_axis_label='sin(x)')

# Add the line plot
p.line(x, y, line_width=2)

# Add grid lines
p.xgrid.grid_line_color = "gray"
p.ygrid.grid_line_color = "gray"

# Show the plot
show(p)
```

19 Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical Bar Chart'.

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, HoverTool
import numpy as np
import pandas as pd

# Prepare the output for Jupyter Notebook (remove if using a script)
output_notebook()

# Generate random categorical data
categories = ['Jeans', 'Shirts', 'Cardigan', 'Suits' ,'warmers']
values = np.random.randint(1, 100, size=len(categories))

# Create a DataFrame for better handling of data
data = pd.DataFrame({'Categories': categories, 'Values': values})

# Create a ColumnDataSource
source = ColumnDataSource(data)

# Create the bar chart
p = figure(x_range=categories, title="Random Categorical Bar Chart",
           x_axis_label='Categories', y_axis_label='Values', tools="")

# Add bars to the plot
bars = p.vbar(x='Categories', top='Values', source=source,
              line_color='green', fill_color='pink')

# Add hover tool
hover = HoverTool()
hover.tooltips = [("Category", "@Categories"), ("Value", "@Values")]
p.add_tools(hover)

# Show the plot
show(p)
```


20 Using Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as 'Simple Line Plot

```python
import plotly.express as px

# Generate random data
np.random.seed(75)
x_values = np.arange(18)
y_values = np.random.randint(10, 200, size=len(x_values))

# Create a DataFrame
df = pd.DataFrame({'X': x_values, 'Y': y_values})

# Create the line plot
fig = px.line(df, x='X', y='Y', title='Simple Line Plot')
fig.update_xaxes(title_text='X Axis Label')
fig.update_yaxes(title_text='Y Axis Label')

# Show the plot
fig.show()
```

21.Using Plotly, create an interactive pie chart of randomly generated data, add labels and percentages, set
the title as 'Interactive Pie Chart'.

```python


# Generate random data
np.random.seed(42)
labels  = ['Jeans', 'Shirts', 'Cardigan', 'Suits' ,'warmers']
values = np.random.randint(1, 100, size=len(labels))

# Create a DataFrame
data = pd.DataFrame({'Labels': labels, 'Values': values})

# Create the pie chart
fig = px.pie(data,
             values='Values',
             names='Labels',
             title=' Pie Chart',
             labels={'Values': 'Count', 'Labels': 'Categories'}, hole=0.2)  # Optional: adds a hole to create a donut chart

# Update the layout to show percentages
fig.update_traces(textinfo='percent+label')

# Show the figure
fig.show()
```
